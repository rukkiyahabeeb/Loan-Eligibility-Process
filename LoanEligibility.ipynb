```python
age=int(input("Enter your age:"))
monthly_income=int(input("Enter your monthly income:"))
credit_score=int(input("Enter your credit score:"))
employment=input("Enter your employment status- Employed/Unemployed:")
if age<21:
    print("Loan Rejected: Age criteria not met")
elif monthly_income<20000:
    print("Loan Rejected: Insufficient income")
elif credit_score<650:
    print("Loan Rejected: Low credit score")
elif employment=="Unemployed":
    print("Loan Rejected: Employment required")
else:
    print("Loan Approved")
    if credit_score>=750:
        print("Premium Loan")
    else:
        print("Standard loan")
```
